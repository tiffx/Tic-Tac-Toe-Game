```python
'''
This is a Tic Tac Toe game is only operated using the num pad. The numbers on the num pad correlates to the position on the \
Tic Tac Toe board.
7 | 8 | 9
4 | 5 | 6
1 | 2 | 3
In order to win, the player must get their marker 3 in a row either horizontal, vertical, or diagonal. The round ends when \
someone wins or there is a tie. When the round ends the player has the option to play again or end the game.
'''
from IPython.display import clear_output
import random

'''
Output for choosing_marker = (Player 1 marker, Player 2 marker)
We have the output as tuple so we can unpack the values later on.
EX:
player1_marker, player2_marker = choosing_marker() ---> tuple unpacking
'''  
    
def choosing_marker(): #player choosing if they want to be 'X' or 'O'
    marker = ''
    while marker.lower() != 'x' and 'o':
        marker = input('Player 1, choose X or O: ')
        if marker.lower() == 'x':
            return ('x','o')
        else:
            return ('o','x')
    
def play_again(): # ask if player wants to play again
    ans = int(input('Do you want to play again? Y=1 or N=2 (use 1 or 2 for your answer)? '))
    if ans == 2:
        return False

def disp_board(board): # outputs display of board
    
    clear_output() # clears old ouput and updates to the new output with the new value/player's move
    
    print('  |' + '   |')
    print(board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('  |' + '   |')
    print('----------')
    print('  |' + '   |')
    print(board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('  |' + '   |')
    print('----------')
    print('  |' + '   |')
    print(board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('  |' + '   |')

def full_board(board): # checks if the board is full
    for n in board:
        if n == ' ':
            return False
    return True
    
def place_move(marker,location,board): # player placing their move
    board[location] = marker

def check_win(marker,board): # check for wins by returning True
    return board[7:10:1] == [marker]*3 or board[4:7:1] == [marker]*3 or board[1:4:1] == [marker]*3 \
or board[1:8:3] == [marker]*3 or board[2:9:3] == [marker]*3 or board[3:10:3] == [marker]*3 \
or board[1:10:4] == [marker]*3 or board[3:8:2] == [marker]*3

def random_first(): # choosing which player goes first
    value = random.randint(0,1)
    if value == 0:
        print('Player 1 make your move.')
        return True
    else:
        print('Player 2 make your move.')
        return False

######################## Game Setup #####################################

print('Welcome to the Tic Tac Toe Game!!!')

play = int(input('Ready to play? Y=1 or N=2 (use 1 or 2 for your answer)?'))

if play == 1:
    game_on = True
else:
    print('Goodbye!!!')

player1_marker, player2_marker = choosing_marker()

print('Player 1: ' + player1_marker)
print('Player 2: ' + player2_marker)

if random_first() == True:
    player1 = True
    player2 = False
else:
    player2 = True
    player1 = False
    
game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']# element at game_board[0] is neglected because we are not using it
disp_board(game_board)

while game_on:

    #### Player 1's move
    while player1:
        
        while not full_board(game_board):
            move = int(input('Player 1 make your move from 1-9: '))
            if game_board[move] == ' ':
                place_move(player1_marker,move,game_board)
                break
            else:
                print('Uh oh space is already taken, please try again.')
                print('\n')
        if full_board(game_board) == True:
            break
        player1 = False
        player2 = True
    
    disp_board(game_board)    
    
    if check_win(player1_marker,game_board) == True:
        print('Player 1 won!!')
        if play_again() == False:
            print('Goodbye!')
            game_on = False
            break
        else:
            game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']# element at game_board[0] is neglected because we are not using it
            disp_board(game_board)
    
    ####### Player 2's move
    while player2:
        while not full_board(game_board):
            move = int(input('Player 2 make your move from 1-9: '))
            if game_board[move] == ' ':
                place_move(player2_marker,move,game_board)
                break
            else:
                print('Uh oh space is already taken, please try again.')
                print('\n')
        if full_board(game_board) == True:
            break
        player1 = True
        player2 = False
    
    disp_board(game_board) 
    
    if check_win(player2_marker,game_board) == True:
        print('Player 2 won!!')
        if play_again() == False:
            print('Goodbye!')
            game_on = False
        else:
            game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']# element at game_board[0] is neglected because we are not using it
            disp_board(game_board)
    
    
    if full_board(game_board) == True:
        print('Tie game!')
        if play_again() == False:
            game_on = False
            print('Goodbye!')
        else:
            game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']# element at game_board[0] is neglected because we are not using it
            disp_board(game_board)
```
